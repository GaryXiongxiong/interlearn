## Chapter 1: Graph Representation

Before implementing DFS, we need a data structure to represent graphs.

### Adjacency List

An **adjacency list** represents a graph as a dictionary (or array) where each key maps to a list of its neighbors. This is the most common representation for DFS because it naturally supports iterating over outgoing edges.

For example, this undirected graph:

```
    A --- B
    |     |
    |     |
    C --- D
```

is represented as:
- `A: [B, C]`
- `B: [A, D]`
- `C: [A, D]`
- `D: [B, C]`

### Adjacency Matrix

An **adjacency matrix** is a 2D array where `matrix[i][j] = 1` if there is an edge from vertex `i` to vertex `j`. It uses O(V^2) space but allows O(1) edge lookups. For DFS on sparse graphs, adjacency lists are preferred since iteration over neighbors takes time proportional to the degree rather than V.

### Key Properties

- **Time complexity of DFS**: O(|V| + |E|) — linear in graph size
- **Space complexity**: O(|V|) for visited set + recursion stack / explicit stack
- DFS explores as far as possible along each branch before backtracking

## Chapter 2: Recursive Depth-First Search

The most intuitive way to implement DFS is **recursion**. The idea is simple:

1. Mark the current vertex as discovered (visited)
2. For each neighbor of the current vertex that hasn't been visited yet, recursively call DFS on it
3. When all neighbors have been explored, return (backtrack)

### How Recursion Naturally Creates a Stack

Each recursive call pushes a new frame onto the call stack. When we reach a dead end (no unvisited neighbors), the function returns — popping that frame — and control flows back to the previous call, which continues with its next neighbor. This is exactly how an explicit stack would work in the iterative version.

### Pseudocode

```
procedure DFS(G, v):
    label v as discovered
    for all directed edges from v to w in G.adjacentEdges(v) do:
        if vertex w is not labeled as discovered then:
            recursively call DFS(G, w)
```

## Chapter 3: Iterative Depth-First Search

## Chapter 3: Iterative Depth-First Search

While recursion is elegant, it has a practical limitation: Python's default recursion limit (~1000 frames). For large or deep graphs, an **iterative** version using an explicit stack avoids stack overflow.

### Key Difference from BFS

DFS and BFS share the same skeleton but differ in one crucial detail:

| Feature | DFS (stack) | BFS (queue) |
|---------|-------------|-------------|
| Data structure | Stack (LIFO) | Queue (FIFO) |
| Exploration strategy | Go deep first | Explore level by level |

### Non-Recursive Pseudocode (with delayed discovery check)

The Wikipedia article describes a variant where the discovered-check is **delayed** until popping from the stack. This can push duplicate vertices onto the stack, but it's simpler:

```
procedure DFS_iterative(G, v):
    let S be a stack
    S.push(v)
    while S is not empty do:
        v = S.pop()
        if v is not labeled as discovered then:
            label v as discovered
            for all edges from v to w in G.adjacentEdges(v) do:
                S.push(w)
```

### Why the Delayed Check Matters

In this version, a vertex can appear on the stack multiple times before being marked discovered. The first time it's popped and found undiscovered, we process it; subsequent pops skip it. This produces DFS but in **reverse neighbor order** compared to the recursive version (because the last pushed neighbor is the first popped).

To match the recursive traversal exactly, we push neighbors in reverse order so they pop in the original order.

## Chapter 4: DFS Traversal Orderings

## Chapter 4: DFS Traversal Orderings

DFS produces several useful vertex orderings beyond simple discovery order. These are fundamental to many graph algorithms.

### Four Vertex Orderings

| Ordering | Description | Use Case |
|----------|-------------|----------|
| **Preorder** | Vertices listed in the order they were first discovered (entered) | Expression trees -> Polish notation |
| **Postorder** | Vertices listed in the order they were last visited (exited/finished) | Expression trees -> Reverse Polish notation |
| **Reverse Preorder** | Reverse of preorder | Rarely used directly |
| **Reverse Postorder** | Reverse of postorder | Topological sorting of DAGs |

### Timestamps: Discovery and Finish Times

We can track two timestamps for each vertex during DFS:
- `discovery_time[v]`: when v is first discovered (entered)
- `finish_time[v]`: when all descendants of v have been fully explored (exited)

These timestamps help classify edges in directed graphs:
- **Tree edge**: w is discovered via v (`discovery[w] > finish[v]`)
- **Back edge**: w is an ancestor of v (`discovery[w] < discovery[v]` and `finish[w] > finish[v]`) — indicates a cycle!
- **Forward edge**: w is a descendant but not a child (in directed graphs)
- **Cross edge**: w is neither ancestor nor descendant

## Chapter 5: Finding Connected Components

A **connected component** of an undirected graph is a maximal set of vertices where every pair has a path between them. DFS provides a simple and efficient way to find all connected components.

### Algorithm

```

for each vertex v in the graph:
    if v not in visited_all:
        component = dfs_collect(v)  # collect all reachable vertices
        components.append(component)
        visited_all.update(component)
```

Each time we start DFS from an unvisited vertex, we discover a new connected component. The algorithm runs in O(|V| + |E|) time — the same as a single DFS traversal.

### Why This Works

## Chapter 6: Cycle Detection and Topological Sorting

### Cycle Detection in Directed Graphs

In a directed graph, DFS classifies edges into four types based on discovery/finish times. A cycle exists if and only if there is at least one **back edge** — an edge from a vertex to one of its ancestors in the DFS tree (i.e., `w` is already discovered but not yet finished when we encounter it).

```
has_back_edge = False
for each neighbor w of v:
    if w is currently on the recursion stack:  # discovered but not finished
        has_back_edge = True  # found a cycle!
```

## Chapter 7: Maze Generation with Randomized DFS

## Chapter 7: Maze Generation with Randomized DFS

DFS isn't just for analysis — it's also a powerful tool for **procedural generation**. A randomized version of DFS can generate mazes with beautiful properties.

### How It Works

The maze is a grid where each cell has walls on all four sides. The algorithm:

1. Start at a random cell, mark it as visited
2. While there are unvisited neighbors:
   - Pick a **random** unvisited neighbor
   - Remove the wall between current cell and chosen neighbor
   - Move to that neighbor (recurse / push to stack)
3. If no unvisited neighbors remain, backtrack

### Mazes from DFS Have These Properties

- **Perfect maze**: Every cell is reachable from every other cell (exactly one path between any two cells)
- **Long corridors**: Because DFS goes deep before backtracking, mazes tend to have long winding paths rather than open areas
- **Single solution path** (between any pair): No loops in the underlying spanning tree

This is sometimes called a "recursive backtracker" maze generator. It's essentially building a **random spanning tree** of the grid graph using DFS.
